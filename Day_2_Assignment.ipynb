```python
import nltk
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
from nltk.corpus import words as nltk_words
from nltk.stem import WordNetLemmatizer
from spellchecker import SpellChecker
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
import re

nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('words')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
```

```python
stop_words = set(stopwords.words('english'))

negations_en = {
    "no", "not", "nor", "never", "neither", "none", "nobody", "nothing", "nowhere",
    "cannot", "can't", "dont", "don't", "doesnt", "doesn't", "didnt", "didn't",
    "isnt", "isn't", "wasnt", "wasn't", "werent", "weren't",
    "havent", "haven't", "hasnt", "hasn't", "hadnt", "hadn't",
    "won", "won't", "wouldnt", "wouldn't", "shouldnt", "shouldn't", "mustnt", "mustn't",
    "shan", "shan't", "mightn", "mightn't", "needn", "needn't",
    "aren", "aren't", "ain", "ain't", "wasn", "wasn't", "weren", "weren't",
    "couldn", "couldn't","didn", "doesn", "don", "hadn", "hasn", "haven", "isn",
    "mustn", "shouldn", "wouldn"
}


stop_words = stop_words-negations_en
```

```python
stop_words
```

```python
spell = SpellChecker()
valid_words = set(w.lower() for w in nltk_words.words())
```

```python
def clean_word(word):
    word = word.lower()
    word = re.sub(r'(.)\1{2,}', r'\1\1', word)
    corrected = spell.correction(word)
    if corrected and corrected.lower() in valid_words:
        return corrected.lower()
    else:
        return ""

def Clean_text(text):
  text = re.sub('[^a-zA-Z]',' ',text)
  text = text.lower()
  text = text.split()

  text = [clean_word(w) for w in text]
  text = [w for w in text if w]

  text = [lemmatizer.lemmatize(word) for word in text if (word not in stop_words and len(word)>1)]
  text = ' '.join(text)
  return text

data['Cleaned_Review'] = data['Review Text'].apply(Clean_text)
```
